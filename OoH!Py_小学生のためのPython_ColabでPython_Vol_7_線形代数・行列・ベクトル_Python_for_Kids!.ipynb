```python
#@title

%matplotlib inline
import matplotlib.patches as pat
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
#自分で追加
from IPython.display import HTML

# グラフ領域作成
fig, ax = plt.subplots(figsize=(8, 8))

#初期設定
def initialize():
  ax.set_xlim(0, 10)
  ax.set_ylim(0, 10)
  ax.set_xticks(np.arange(0, 10, 1))
  ax.set_yticks(np.arange(0, 10, 1))
  ax.grid()

initialize()

#初期ベクトル設定
X = 1
Y = 7
s = np.array([X, Y])
U = 2
V = 2
v0 = np.array([U, V])
Q = ax.quiver(X, Y, U, V, color='cyan', width=0.1,  headwidth=1, headlength=0.5,
              angles='xy', scale_units='xy', scale=1)
plt.show()

#ベクトル作成関数
def Quiver(X,Y,U,V):
  Q = ax.quiver(X, Y, U, V, color='cyan', width=0.1,  headwidth=1, headlength=0.5,
                angles='xy', scale_units='xy', scale=1)
  return Q

#up-date関数
def update(frame, Q, X, Y):
  ax.cla()
  initialize()

  if (0<= frame <=121):
    X = X + frame*0.05
    Y = Y
    if 20< frame <=70:
# 縮小行列c 対称がベクトルなので縮小ではなく短縮になる
      c = np.array([[0.5, 0],
                    [0, 0.5]])
      v1 = c @ v0
      U = v1[0]
      V = v1[1]

    else:
      U = v0[0]
      V = v0[1]

    Q.set_UVC(U,V)
    Quiver(X,Y,U,V)

#180度回転，回転行列Rotate(反時計回り)
  elif 121<= frame <=302:
    X = 7
    Y = 7
    sind = np.sin(np.deg2rad(frame-121))
    cosd = np.cos(np.deg2rad(frame-121))
    Rotate = np.array([[cosd, -sind],
                       [sind, cosd]])
    vr = Rotate @ v0
    U = vr[0]
    V = vr[1]
    Q.set_UVC(U,V)
    Quiver(X,Y,U,V)

#(x, Y) = (7, 7)から(5, 5)まで直進
  elif 302< frame <=343:
    X = 7 - (frame - 302) * 0.05
    Y = 7 - (frame - 302) * 0.05
    U = -2
    V = -2
    Quiver(X,Y,U,V)

#2回転，回転行列Rotate2(時計回り)
  elif 343<= frame <=525:
    X = 5
    Y = 5
    U = -2
    V = -2
    vr = np.array([U, V])
    sind = np.sin(np.deg2rad(frame-344) * 4)
    cosd = np.cos(np.deg2rad(frame-344) * 4)
    Rotate2 = np.array([[cosd, sind],
                       [-sind, cosd]])
    vr2 = Rotate2 @ vr
    U = vr2[0]
    V = vr2[1]
    Q.set_UVC(U,V)
    Quiver(X,Y,U,V)

  elif 525< frame <=566:
    X = 5 - (frame - 525) * 0.05
    Y = 5 - (frame - 525) * 0.05
    U = -2
    V = -2
    Quiver(X,Y,U,V)

#(2, 2)を中心に180度回転＝点対称移動，回転行列Rotate3(反時計回り)
  elif 566< frame <=627:
    X = 3
    Y = 3
    U = -2
    V = -2
    vrxy = np.array([X, Y]) - np.array([2, 2])
    vruv = np.array([U, V])
    sind = np.sin(np.deg2rad(frame-566) * 3)
    cosd = np.cos(np.deg2rad(frame-566) * 3)
    Rotate3 = np.array([[cosd, -sind],
                        [sind, cosd]])
    #原点を中心に回転させてから元に戻す
    vrxy2 = Rotate3 @ vrxy + np.array([2, 2])
    X = vrxy2[0]
    Y = vrxy2[1]
    vruv2 = Rotate3 @ vruv
    U = vruv2[0]
    V = vruv2[1]
    Q.set_UVC(U,V)
    Quiver(X,Y,U,V)

#(X, Y) = (7, 1)まで平行移動
  elif 627< frame <=648:
    X, Y = 1, 1
    U, V = 2, 2
    X = X + (frame - 627) * 0.3
    Y = Y
    Quiver(X,Y,U,V)

#(6, 4)を中心に180度回転，回転行列Rotate4(反時計回り)
  elif 648< frame <=709:
    X = 7
    Y = 1
    U = 2
    V = 2
    vrxy = np.array([X, Y]) - np.array([6, 4])
    vruv = np.array([U, V])
    sind = np.sin(np.deg2rad(frame-648) * 3)
    cosd = np.cos(np.deg2rad(frame-648) * 3)
    Rotate4 = np.array([[cosd, -sind],
                        [sind, cosd]])
    #原点を中心に回転させてから元に戻す
    vrxy2 = Rotate4 @ vrxy + np.array([6, 4])
    X = vrxy2[0]
    Y = vrxy2[1]
    vruv2 = Rotate4 @ vruv
    U = vruv2[0]
    V = vruv2[1]
    Q.set_UVC(U,V)
    Quiver(X,Y,U,V)

#直線y＝xに対しての線対称移動
  elif 709< frame <=740:
    #y＝xのグラフの描画
    x = np.arange(0, 10, 0.1)
    y = x
    ax.plot(x, y)
    ax.quiver(4.5, 5.5, 1, -1.25, color="black", width=0.003, headwidth=7,
              angles='xy', scale_units='xy', scale=1)
    ax.quiver(5, 7, -2, -2, color='cyan', width=0.1,  headwidth=1, headlength=0.5,
                  angles='xy', scale_units='xy', scale=1)
    if 719< frame <=740:
      X = 5
      Y = 7
      U = -2
      V = -2
      vmxy = np.array([X, Y])
      #線対称移動用Mirror行列
      Mirror = np.array([[0, 1],
                         [1, 0]])
      #線対称移動
      vmxy2 = Mirror @ vmxy
      X = vmxy2[0]
      Y = vmxy2[1]
      Quiver(X,Y,U,V)


#アニメーション作成
ani = animation.FuncAnimation(fig, update, fargs=(Q,X,Y), interval=50, frames=740, repeat=False)

# Display方法その1
HTML(ani.to_html5_video())

# Display方法その2(こちらにする場合は，上のHTML()は # にする)
#html = ani.to_jshtml()
#HTML(html)
```

```python
#@title
# 実行時間約1分

%matplotlib inline
import matplotlib.patches as pat
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
from IPython.display import HTML
import math

# グラフ領域作成
fig, ax = plt.subplots(figsize=(8, 8))

#数値設定
root3 = math.sqrt(3)

#初期設定
def initialize():
  ax.set_xlim(0, 10)
  ax.set_ylim(0, 10)
  ax.set_xticks(np.arange(0, 10, 1))
  ax.set_yticks(np.arange(0, 10, 1))
  ax.grid()

initialize()

#腕1ベクトル作成関数
def Quiver_A1(x,y,u,v):
  P = ax.quiver(x, y, u, v, color='green', width=0.01,  headwidth=0, headlength=0,
                headaxislength=0, angles='xy', scale_units='xy', scale=1)
  return P

#腕2ベクトル作成関数
def Quiver_A2(X,Y,U,V):
  Q = ax.quiver(X, Y, U, V, color='cyan', width=0.01, headwidth=8, headlength=2,
                headaxislength=6, angles='xy', scale_units='xy', scale=1)
  return Q

#回転行列(反時計回り)
def Rotate_Inv(t, z, w):
  sind = np.sin(np.deg2rad(t - z) * w)
  cosd = np.cos(np.deg2rad(t - z) * w)
  RI = np.array([[cosd, -sind],
                 [sind, cosd]])
  return RI

#回転行列(時計回り，4倍速)
def Rotate(t, z, w):
  sind = np.sin(np.deg2rad(t - z) * w)
  cosd = np.cos(np.deg2rad(t - z) * w)
  R = np.array([[cosd, sind],
                [-sind, cosd]])
  return R

#縮小行列(腕2だけ適用)
def Short(t, z):
  Short = np.array([[1-0.035*(t-z), 0],
                    [0, 1-0.035*(t-z)]])
  return Short


#拡大行列
def Long(t, z):
  Long = np.array([[1+0.05*(t-z), 0],
                   [0, 1+0.05*(t-z)]])
  return Long


#初期ベクトル設定
#腕1(Arm1)ベクトル
x = 1
y = 4
u = 3
v = 0
P = Quiver_A1(x,y,u,v)

#腕2(Arm2)ベクトル
X = 4
Y = 4
U = 2.5
V = 0
Q = Quiver_A2(X,Y,U,V)

plt.show()

#腕1,2の動き設定
def Arms(t):
  if 0<= t <= 30:
    z = 0
    w = 1
    x,y,u,v,X,Y,U,V = 1,4,3,0,4,4,2.5,0

    vect = np.array([u, v])
    RI = Rotate_Inv(t, z, w)
    vect = RI @ vect
    u = vect[0]
    v = vect[1]
    P.set_UVC(u,v)
    Quiver_A1(x,y,u,v)

    S = np.array([X, Y]) - np.array([1, 4])
    S = RI @ S + np.array([1, 4])
    X = S[0]
    Y = S[1]
    Vect = np.array([U, V])
    Vect = RI @ Vect
    U = Vect[0]
    V = Vect[1]
    Q.set_UVC(U,V)
    Quiver_A2(X,Y,U,V)

  elif 30<= t <=60:
    z = 30
    w = 1
    x,y,u,v,X,Y,U,V = 1,4,1.5*root3,1.5,1+1.5*root3,5.5,1.25*root3,1.25

    vect = np.array([u, v])
    R = Rotate(t, z, w)
    vect = R @ vect
    u = vect[0]
    v = vect[1]
    P.set_UVC(u,v)
    Quiver_A1(x,y,u,v)

    S = np.array([X, Y]) - np.array([1, 4])
    S = R @ S + np.array([1, 4])
    X = S[0]
    Y = S[1]
    Vect = np.array([U, V])
    Vect = R @ Vect
    U = Vect[0]
    V = Vect[1]
    Q.set_UVC(U,V)
    Quiver_A2(X,Y,U,V)

  elif 60<= t <=75:
    z = 60
    x,y,u,v,X,Y,U,V = 1,4,3,0,4,4,2.5,0

    Quiver_A1(x,y,u,v)

    Vect = np.array([U, V])
    L = Long(t, z)
    Vect = L @ Vect
    U = Vect[0]
    V = Vect[1]
    Q.set_UVC(U,V)
    Quiver_A2(X,Y,U,V)

  elif 75<= t <=90:
    z = 75
    x,y,u,v,X,Y,U,V = 1,4,3,0,4,4,4.75,0

    Quiver_A1(x,y,u,v)

    Vect = np.array([U, V])
    S = Short(t, z)
    Vect = S @ Vect
    U = Vect[0]
    V = Vect[1]
    Q.set_UVC(U,V)
    Quiver_A2(X,Y,U,V)

  elif 90<= t <=180:
    z = 90
    x,y,u,v,X,Y,U,V = 1,4,3,0,4,4,2.5,0

    Quiver_A1(x,y,u,v)

    Vect = np.array([U, V])
    w = 4
    R = Rotate(t, z, w)
    Vect = R @ Vect
    U = Vect[0]
    V = Vect[1]
    Q.set_UVC(U,V)
    Quiver_A2(X,Y,U,V)

  elif 180<= t <= 211:
    z = 180
    w = 1
    x,y,u,v,X,Y,U,V = 1,4,3,0,4,4,2.5,0

    Quiver_A1(x,y,u,v)

    RI = Rotate_Inv(t, z, w)
    Vect = np.array([U, V])
    Vect = RI @ Vect
    U = Vect[0]
    V = Vect[1]
    Q.set_UVC(U,V)
    Quiver_A2(X,Y,U,V)
  #腕1，2が同時に30度反時計回り
  elif 211<= t <= 241:
    z = 211
    w = 1
    x,y,u,v,X,Y,U,V = 1,4,3,0,4,4,1.25*root3,1.25

    vect = np.array([u, v])
    RI = Rotate_Inv(t, z, w)
    vect = RI @ vect
    u = vect[0]
    v = vect[1]
    P.set_UVC(u,v)
    Quiver_A1(x,y,u,v)

    S = np.array([X, Y]) - np.array([1, 4])
    S = RI @ S + np.array([1, 4])
    X = S[0]
    Y = S[1]
    Vect = np.array([U, V])
    Vect = RI @ Vect
    U = Vect[0]
    V = Vect[1]
    Q.set_UVC(U,V)
    Quiver_A2(X,Y,U,V)
  #腕2が伸びながら330度反時計回り(10倍速)
  elif 241<= t <= 275:
    z = 241
    w = 10
    x,y,u,v,X,Y,U,V = 1,4,1.5*root3,1.5,1+1.5*root3,5.5,1.25,1.25*root3

    Quiver_A1(x,y,u,v)

    #伸びながら330度反時計回り回転
    RI = Rotate_Inv(t, z, w)
    L = Long(t, z)
    Vect = np.array([U, V])
    Vect = RI @ L @ Vect
    U = Vect[0]
    V = Vect[1]
    Q.set_UVC(U,V)
    Quiver_A2(X,Y,U,V)


#up-date関数
def update(frame):
  ax.cla()
  initialize()

  t = frame

  Arms(t)


#アニメーション作成
#ani = animation.FuncAnimation(fig, update, fargs=(P, x, y, Q, X, Y), interval=200, frames=61, repeat=False)
ani = animation.FuncAnimation(fig, update, interval=200, frames=275, repeat=False)
HTML(ani.to_html5_video())

```

#★ベクトル

＜参考＞[Cognicull\ベクトル](https://cognicull.com/ja/oczgnusb)

ベクトルVectorは，「**向きと大きさを持つ量**」のことで，

速度，力，風(風向)などはベクトルです。

速度は「速さ＋向き」で，大きさと向きがありますが，

速さは大きさを示すだけの量なのでベクトルではありません。

ある点が動く場合，「どの方向にどれだけ位置を変えたか」(**変位**(へんい)といいます)も

ベクトルです。

\

ベクトルを表すには「**矢印**やじるし(矢線・やせん)」を使います。

ベクトルは上で紹介した`quiver()`で作成します

矢印--->のおしり(左側)を`始点`，先端(右側）を`終点`といいます

下の黒いベクトルは，点(1, 2)を始点として，右に3，上に2の向きになっていますから，

`quiver(x, y, u, v)`=`quiver(1, 2, 3, 2)`と表します

上でu, v は向き(方向)を表すといったのは，このことです

\

赤いベクトル，点(7, 8)を始点として，向きが，左に5,下に2なので，

`quiver(x, y, u, v)`=`quiver(7, 8, -5, -2)`となります。

\

向き(方向)では，`右，上は$+$(プラス)`で，`左，下は$-$(マイナス)`です。

\

なお，u,v（の絶対値）でベクトルの大きさも分かります

#★ ベクトルの平行移動アニメーション


FuncAnimationのコードの基本的な流れは次の通りです

1．必要モジュールの読み込み (import )

2．描画領域の作成 (fig, ax = plt.subplots())

3．描画の基本設定＝アニメーション化するグラフ(ここではベクトル）の用意 

4．`update()`関数(描画更新用関数のこと）作成

  (def update(frame,…):

5．アニメーション化 (anim = FuncAnimation(fig, update))



update()関数の引数に`frame`がありますが，名前はframeではなく，iやxなどなんでもOK

これは，0,1,2,3,4$\dots$と増えていく整数です

frameの値が１ずつ増加することを利用して，描画を更新する関数を作成します

`frame`フレームは指定がないときは，0.2秒(200ミリ秒)ごとに切り替わります

フレームの更新間隔を指定するのが`interval引数`で，

指定しないと既定値(デフォルト)は200で，単位はミリ秒


ベクトルの移動アニメーションのコードに説明を入れます

```python

# %matplotlib inline  Jupiter Notebookを使う場合に入れる(Colabでは不要）
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
# 動画作成のときに必要
from IPython.display import HTML

# グラフ領域作成
fig, ax = plt.subplots(figsize=(8,4))

# ベクトルの用意(初期設定)
# 始点(起点)
s = np.array([1, 1])
# 方法(向き)ベクトル
a = np.array([5, 2])

# up-date関数(グラフ更新関数)
def update(frame):
  # 描画をいったん消す
  ax.cla()
  # 軸などの表示
  ax.set_xlim(0, 100)
  ax.set_ylim(0, 4)
  ax.grid()
  # ベクトル描画
  # ベクトルを真横に移動させるので，始点のx座標s[0]だけが変わるので，それをframeとして
  # 1フレーム(0.2秒)ごとに動かす  始点のy座標s[1]，方向ベクトルは初期設定のまま  
  ax.quiver(frame, s[1], a[0], a[1], angles='xy', scale_units='xy', scale=1,
            color="black", width=0.005, headwidth=1, headlength=1)

#アニメーション作成
# FuncAnimation()関数では，figとupdateは必ず必要な引数
# fig = figure(アニメを描く）の意味で，updateは上の更新関数で，defで定義した名前を入れる
ani = animation.FuncAnimation(fig, update)
# aniをHTML()関数で動かす
HTML(ani.to_html5_video())
```

速く動かしたいときは，FuncAnimation()関数の引数で，

`interval = 50` (既定値(デフォルト)は200)

interval / 1000 * frames = アニメーションの時間

くり返さないようにするには，

`repeat = False`  (デフォルトはTrue)



```python

import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
from IPython.display import HTML

fig, ax = plt.subplots(figsize=(8,4))

s = np.array([1, 1])
a = np.array([5, 2])

def update(frame):
  ax.cla()
  ax.set_xlim(0, 100)
  ax.set_ylim(0, 4)
  ax.grid()
  ax.quiver(frame, s[1], a[0], a[1], angles='xy', scale_units='xy', scale=1,
            color="black", width=0.005, headwidth=1, headlength=1)

# interval=50 --> 時間が20秒から5秒くらいになる
# repeat=False --> アニメーションが1回で止まる
ani = animation.FuncAnimation(fig, update, interval=50, repeat=False)

HTML(ani.to_html5_video())
```

```python

import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
from IPython.display import HTML

fig, ax = plt.subplots(figsize=(6,6))

# ベクトルの始点 ななめ移動なので，x,y座標が変化するのでそれぞれの座標をX,Yに入れる
s = np.array([2, 5])
X = s[0]
Y = s[1]
a = np.array([2, 3])

# x,y座標を少しずつ動かす，x座標は右へ(+), y座標は下へ(-)
def update(frame, X, Y):
  ax.cla()
  ax.set_xlim(0, 9)
  ax.set_ylim(0, 9)
  ax.grid()
  X = s[0] + frame*0.05
  Y = s[1] - frame*0.05
  ax.quiver(X, Y, a[0], a[1], angles='xy', scale_units='xy', scale=1,
            color="black", width=0.005, headwidth=1, headlength=1)

# fargs :	update()関数に渡すframe以外の追加の引数で，タプルで渡します
ani = animation.FuncAnimation(fig, update, fargs=(X,Y), interval=50, repeat=False)

html = ani.to_jshtml()
HTML(html)

```

# ★ 行列の説明

---
**<< 行列とは >>**

\

$\begin{pmatrix}1 & 2 \\3 & 4 \end{pmatrix}$とか
 $\begin{bmatrix}
 a & b & c\\
 d & e & f
  \end{bmatrix} $のように数字や文字を長方形にならべ，

$\begin{pmatrix} & \\ & \end{pmatrix}$や$\begin{bmatrix} & \\ & \end{bmatrix} $でくくったものを行列といいます

\

カッコの中の1つ1つの数字や文字を**成分**とか**要素**といいます

成分も要素も英語では**Element(エレメント）**

\

行列はまとめて`A`,`B`,`X`などの大文字で表します

\
行列Aは，
A = $\begin{pmatrix}1 & 2 \\3 & 4 \end{pmatrix}$


行列Bは，
B = $\begin{bmatrix} a & b & c\\ d & e & f  \end{bmatrix} $

\

---

ベクトルも，$(3, 5)$ あるいは $\begin{pmatrix}3 \\ 5 \end{pmatrix}$と書きますので

1行2列の行列＝(1$\times$2の行列)

とか

2行1列の行列＝(2$\times$1の行列)

になります

---

\
行列の横方向の並びを**行(Rowロウ)**、縦方向の並びを**列(Columnコラム)**と言います。

行列Aを**2行2列の行列(2$\times$2行列)**といい，

$\begin{pmatrix}1 & 2 \\ &  \end{pmatrix}$を**第1行**，$\begin{pmatrix} & \\3 & 4 \end{pmatrix}$を**第2行**

$\begin{pmatrix}1 &  \\ 3 &  \end{pmatrix}$を**第1列**，$\begin{pmatrix} & 2 \\ & 4 \end{pmatrix}$を**第2列**

\
行列Bは**2行3列の行列(2$\times$3行列)**です




---

ベクトルを拡大・縮小するには，ベクトルに行列をかけるといいましたが，

行列のかけ算の記号は`@`を使います

ベクトルvと行列Bのかけ算は，上で言いましたが，

**(行列) $\times$ (ベクトル)**なので，

`B @ v`

です

別のかき方で，numpyの`dot()`関数を使い

`np.dot(B, v)`

でも同じです

---




================================================================\
(参考)

拡大縮小，回転だけでなく平行移動もできる行列がありますがこれは次のステップで

やってください

たとえば，x軸方向に+4，y軸方向に-3平行移動させる行列は

 $\begin{pmatrix}
 1 & 0 & 4\\
 0 & 1 & -3\\
 0 & 0 & 1
  \end{pmatrix} $

となります

3$\times$3行列なので平面上の点(x, y)を平行移動させるには

$\begin{pmatrix} 1 & 0 & 4\\ 0 & 1 & -3\\ 0 & 0 & 1\end{pmatrix} $$\begin{pmatrix} x\\ y\\ 1\end{pmatrix} $

としてかけ算をします




---
**＜ベクトルの拡大・縮小＞**

ベクトルをk倍に拡大。縮小する行列は，2$\times$2行列の

$\begin{pmatrix}k & 0 \\0 & k \end{pmatrix}$

です。

kが**1より大きければ拡大**，kが**0から1の間であれば縮小**です

\

この行列もnumpyの`array()`関数を使って

    np.array([[k, 0],
              [0, k]])
と書きます

1行で

    np.array([[k, 0],[0, k]])
でもいいのですが，行列っぽく見えるような上のかき方をします

`[  ]`が2重になっていることに注意で，これがタテ$\times$横の2次元の配列

であることが分かります

\
ちなみに，3$\times$2行列Cを，
C = $\begin{pmatrix} 1 & 2 \\ 3 & 4\\ 5 & 6  \end{pmatrix} $ とすると，

    C ＝ np.array([[1, 2],
                   [3, 4],
                   [5, 6]])
となります

\

ベクトルv = np.array([4, 3])に行列Bをかけて2倍にするには

    行列B = np.array([[2, 0],
                      [0, 2]])
として，

    B @ v
または

    np.dot(B, v)

と書きます


================================================================\
(参考)

行列Bをベクトルvにかけるかけ算は

ベクトルv = np.array([4, 3])として，

    B @ v

でしたが，これを数学の行列の式にすると，

$\begin{pmatrix} 2 & 0\\ 0 & 2  \end{pmatrix} \begin{pmatrix} 4 \\ 3  \end{pmatrix} = \begin{pmatrix} 8 \\ 6  \end{pmatrix}$

となります

\

行列のかけ算については，「★ 補足」の[行列のかけ算](#行列のかけ算-id)で説明してます


---
以上をもとにベクトル`v0`の拡大・縮小のコードが以下になります

ベクトルが重ならないように，拡大・縮小後に始点のy座標a[1]を動かしてから表示してます

```python

import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6, 6))

ax.set_xlim(0, 20)
ax.set_ylim(0, 20)
plt.xticks(np.arange(0, 20, 1)) 
plt.yticks(np.arange(0, 20, 1)) 

ax.grid()

# ベクトルの始点
a = np.array([5, 10])
# ベクトルv0
v0 = np.array([4, 3])

# 黒ベクトル
ax.quiver(a[0], a[1], v0[0], v0[1], color="black", angles='xy', scale_units='xy', scale=1)

# 2倍拡大行列B
B = np.array([[2, 0],
             [0, 2]])

# 拡大変換　
v1 = B @ v0    # または，v1 = np.dot(B, v0)
# 位置をずらす
a[1] += 3
# 赤ベクトル
ax.quiver(a[0], a[1], v1[0], v1[1], color="red", angles='xy', scale_units='xy', scale=1)

# 縮小行列C
C = np.array([[0.5, 0],
             [0, 0.5]])

# 0.5倍縮小変換
v2 = C @ v0     # または, v2 = np.dot(C, v0)
# 位置をずらす
a[1] -= 6
# 緑ベクトル
ax.quiver(a[0], a[1], v2[0], v2[1], color="green", angles='xy', scale_units='xy', scale=1)

plt.show()
```

#★ ベクトルの拡大・縮小アニメーション

アニメーションにしてます

初期設定を`initialize()`という関数にし，

ベクトルを動かし，時間によって拡大・縮小する関数`move_vector()`にしてます

ベクトルの太さも変えたいときは，quiver()の引数`width`を変えます

```python

import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
from IPython.display import HTML

fig, ax = plt.subplots(figsize=(6, 6))

# 初期設定をinitialize()という関数にする
def initialize():
  ax.set_xlim(0, 21)
  ax.set_ylim(0, 101)
  ax.grid()

# ベクトルを動かす関数move_vector()にしてます
def move_vector(t):
  # 始点位置
  a = np.array([5, 80])
  # もとになるベクトル
  v0 = np.array([4, 3])
  # y座標を下に動かす
  a[1] -= t 

  if 10 <= t <= 30:
    # 2倍拡大行列B
    B = np.array([[2, 0],
                  [0, 2]])
    # ベクトル拡大
    v1 = B @ v0
    # 赤ベクトル
    ax.quiver(a[0], a[1], v1[0], v1[1], color="red", width=0.01, angles='xy', scale_units='xy', scale=1)

  elif 40 <= t <= 60:
    # 0.5倍縮小行列C
    C = np.array([[0.5, 0],
                  [0, 0.5]])
    # ベクトル縮小
    v2 = C @ v0
    # 赤ベクトル
    ax.quiver(a[0], a[1], v2[0], v2[1], color="green", width=0.01, angles='xy', scale_units='xy', scale=1)
  
  else:
    ax.quiver(a[0], a[1], v0[0], v0[1], color="black", width=0.01, angles='xy', scale_units='xy', scale=1)

# 更新関数
# frameをtにしてます
def update(t):
    ax.cla()
    initialize()
    move_vector(t)

ani = animation.FuncAnimation(fig, update)
html = ani.to_jshtml()
HTML(html)
```

#★ ベクトルの回転

---
ベクトルを回転させるには，ベクトルに回転行列をかけます

`回転行列R(rotation matrix)`はベクトルの始点を軸に回転させます。

\

$R = \begin{pmatrix}cos\theta & -sin\theta \\sin\theta & cos\theta \end{pmatrix}$

\

この行列もnumpyの`array()`関数を使って

np.array([[cos$\theta$, -sin$\theta$],\
&emsp;&emsp;&emsp;&emsp;[sin$\theta$, cos$\theta$]])

と書きます

\

ここで出てきた$sin\theta$や$cos\theta$を**三角関数(さんかくかんすう)**といいます

三角関数の簡単な説明は下に書きましたが，ここで大切なのは，



1.   ベクトルを回転させるには**回転行列**を使う

     ここでの回転は，`反時計回りが順回転`，`時計回りは逆回転`です
2.   三角関数を使うには`math(マス)`モジュールをインポートする
3.   **$\theta$(シータ)**は回転角度で，角度は小学校で習った°(度)ではなく

     別の角度の表し方(**弧度法・こどほう＝ラジアン**）を使う

     学校で習った角度を**度数法(どすうほう)**といいます

の3点です

\

2.について，はじめのモジュールのインポートで

    import math

が必要です

\
3.について，**mathモジュールの三角関数は，角度の単位としてラジアンを使います**

例えばベクトルを60度(60°）回転させたいときは，角度を°からラジアンに変える関数

がmathモジュールの中にあり，`math.radians()`です

われわれは60°のほうが慣れているので，回転行列を


    R = np.array([[cos90, -sin90],
                  [sin90,  cos90]])

とかきたいので，度をラジアンに変更するために

    sin60 = np.sin(math.radians(60))
    cos60 = np.cos(math.radians(60))

というコードを書いておきます

**かけ算の仕方は，拡大・縮小と同じです**

\
(参考)

度をラジアンに変える関数はnumpyの中にもあり，

sin60 = np.sin(np.deg2rad(60))

cos60 = np.cos(np.deg2rad(60))

となります



下のコードは，黒ベクトルv0を90°回転させて赤ベクトルv1にします

```python

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
# mathモジュールのインポート
import math

fig, ax = plt.subplots(figsize=(6, 6))

ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
plt.xticks(np.arange(0, 10, 1)) #x軸の目盛の間隔
plt.yticks(np.arange(0, 10, 1)) #y軸の目盛の間隔
ax.grid()

# ベクトルの始点
a = np.array([5, 3])
# ベクトル成分
v0 = np.array([4, 4])
# 回転前黒ベクトル
ax.quiver(a[0], a[1], v0[0], v0[1], color="black", angles='xy', scale_units='xy', scale=1)
# ベクトルの先端に名前を表示
ax.text(a[0]+v0[0], a[1]+v0[1], "v0", color="black", size=16)

# 角度法を弧度法に変換
sin90 = np.sin(math.radians(90))
cos90 = np.cos(math.radians(90))
# 回転行列の作成
R = np.array([[cos90, -sin90],
              [sin90,  cos90]])

# 回転して表示
v1 = R @ v0

ax.quiver(a[0], a[1], v1[0], v1[1], color="red", #width=0.005, headwidth=1, headlength=1,
          angles='xy', scale_units='xy', scale=1)
# ベクトルの先端に名前を表示
ax.text(a[0]+v1[0], a[1]+v1[1], "v1", color="red", size=16)

# 動きを示す矢印
Arc = patches.Arc(xy=(5, 5), width=4, height=4, angle=50, theta1=0, theta2=90, color='green')
ax.add_patch(Arc)
Arrow = patches.Arrow(x=3.5, y=6.3, dx=0.1, dy=0.1, width=0.5)
ax.add_patch(Arrow)

plt.show()
```

#★ ベクトルの対称移動

線対称，点対称は知ってる小学生もいると思います

2つの図形があり，

ある直線に関して折り返したときにピッタリ重なる場合を**線対称**な図形

ある点を中心として180°回転させたときにピッタリ重なる場合を**点対称**な図形

といいます


---
＜＜ベクトルの対称移動＞＞

ベクトルに2$\times$2行列をかけます

**かけ算の仕方は，拡大・縮小，回転と同じです**

\

ベクトルを**x軸**に対して(線)対称移動＝折り返しする行列は，2$\times$2行列の

$\begin{pmatrix}1 & 0 \\0 & -1 \end{pmatrix}$


\

この行列もnumpyの`array()`関数を使って

    np.array([[1, 0],
              [0, -1]])
と書きます

\

ベクトルを**y軸**に対して(線)対称移動する行列は，

$\begin{pmatrix}-1 & 0 \\0 & 1 \end{pmatrix}$

\

ベクトルを**直線y＝x**に対して(線)対称移動する行列は，

$\begin{pmatrix}0 & 1 \\1 & 0 \end{pmatrix}$

\

ベクトルを**原点O**に対して(点)対称移動する行列は，

$\begin{pmatrix}-1 & 0 \\0 & -1 \end{pmatrix}$

\
以下が作成コードです




ベクトルを今までの矢印からロケット型にしてます

```python

import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(8, 8))

ax.set_xlim(-8, 8, 1)
ax.set_ylim(-8, 8, 1)
ax.set_xticks(np.arange(-8, 8, 1))
ax.set_yticks(np.arange(-8, 8, 1))
ax.grid()

# x軸y軸作成
ax.annotate(' ', xy = (0, 8), size = 5, xytext = (0, -8),
            arrowprops=dict(width=0.5, headwidth=5, headlength=10))
ax.annotate(' ', xy = (8, 0), size = 5, xytext = (-8, 0),
            arrowprops=dict(width=0.5, headwidth=5, headlength=10))

# 原点,x軸，y軸表示
ax.text(-0.6, -0.6, "O",  fontweight="bold", fontstyle="italic", fontsize=16)
ax.text(7.5, -0.6, "x",  fontweight="bold", fontstyle="italic", fontsize=16)
ax.text(-0.6, 7.5, "y",  fontweight="bold", fontstyle="italic", fontsize=16)


# ベクトル設定
s0 = np.array([4, 2])
v0 = np.array([2, 3])
Q = ax.quiver(s0[0], s0[1], v0[0], v0[1], color='cyan', width=0.1,  headwidth=1, headlength=0.5,
              angles='xy', scale_units='xy', scale=1)


# x軸に対して線対称移動用MirrorX行列
MirrorX = np.array([[1, 0],
                    [0, -1]])
# 線対称移動
# 始点も移動することに注意
sx = MirrorX @ s0
vx = MirrorX @ v0
# 移動後のベクトル
Q = ax.quiver(sx[0], sx[1], vx[0], vx[1], color='green', width=0.1,  headwidth=1, headlength=0.5,
              angles='xy', scale_units='xy', scale=1)

# y軸に対して線対称移動用MirrorY行列
MirrorY = np.array([[-1, 0],
                    [0, 1]])
# 線対称移動
sy = MirrorY @ s0
vy = MirrorY @ v0
# 移動後のベクトル
Q = ax.quiver(sy[0], sy[1], vy[0], vy[1], color='lightblue', width=0.1,  headwidth=1, headlength=0.5,
              angles='xy', scale_units='xy', scale=1)

# 直線y=xに対して線対称移動用MirrorYX行列
MirrorYX = np.array([[0, 1],
                     [1, 0]])
# 線対称移動
syx = MirrorYX @ s0
vyx = MirrorYX @ v0
# 移動後のベクトル
Q = ax.quiver(syx[0], syx[1], vyx[0], vyx[1], color='pink', width=0.1,  headwidth=1, headlength=0.5,
              angles='xy', scale_units='xy', scale=1)

# 直線y＝x
ax.plot([0, 7], [0, 7])


# 原点に対して線対称移動用Symmtry行列(Symmtry=対称の英語)
Symmtry = np.array([[-1, 0],
                    [0, -1]])
# 点対称移動
so = Symmtry @ s0
vo = Symmtry @ v0
# 移動後のベクトル
Q = ax.quiver(so[0], so[1], vo[0], vo[1], color='red', width=0.1,  headwidth=1, headlength=0.5,
              angles='xy', scale_units='xy', scale=1)

plt.show()
```

#★ ベクトルの移動アニメーション

効率が悪いコードのせいか，動くまで2分ほどかかります

直進ー＞縮小ー＞直進ー＞180度回転ー＞ななめ直進ー＞2回転ー＞ななめ直進

ー＞途中で2回転ー＞ななめ直進ー＞(2,2)を中心に180度回転ー＞平行移動

ー＞(6, 4)を中心に180度回転ー＞y＝xに対して線対称移動

という感じで動きます


```python

import matplotlib.patches as pat
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
from IPython.display import HTML

fig, ax = plt.subplots(figsize=(8, 8))

# 初期設定
def initialize():
  ax.set_xlim(0, 10)
  ax.set_ylim(0, 10)
  ax.set_xticks(np.arange(0, 10, 1))
  ax.set_yticks(np.arange(0, 10, 1))
  ax.grid()

initialize()

# 初期ベクトル設定
X = 1
Y = 7
s = np.array([X, Y])
U = 2
V = 2
v0 = np.array([U, V])
Q = ax.quiver(X, Y, U, V, color='cyan', width=0.1,  headwidth=1, headlength=0.5,
              angles='xy', scale_units='xy', scale=1)
plt.show()

# ベクトル作成関数
def Quiver(X,Y,U,V):
  Q = ax.quiver(X, Y, U, V, color='cyan', width=0.1,  headwidth=1, headlength=0.5,
                angles='xy', scale_units='xy', scale=1)
  return Q

# up-date関数
def update(frame, Q, X, Y):
  ax.cla()
  initialize()

  if (0<= frame <=121):
    X = X + frame*0.05
    Y = Y
    if 20< frame <=70:
# 縮小行列c 対称がベクトルなので縮小ではなく短縮になる
      c = np.array([[0.5, 0],
                    [0, 0.5]])
      v1 = c @ v0
      U = v1[0]
      V = v1[1]

    else:
      U = v0[0]
      V = v0[1]

    Q.set_UVC(U,V)
    Quiver(X,Y,U,V)

# 180度回転，回転行列Rotate(反時計回り)
  elif 121<= frame <=302:
    X = 7
    Y = 7
    sind = np.sin(np.deg2rad(frame-121))
    cosd = np.cos(np.deg2rad(frame-121))
    Rotate = np.array([[cosd, -sind],
                       [sind, cosd]])
    vr = Rotate @ v0
    U = vr[0]
    V = vr[1]
    Q.set_UVC(U,V)
    Quiver(X,Y,U,V)

# (x, Y) = (7, 7)から(5, 5)まで直進
  elif 302< frame <=343:
    X = 7 - (frame - 302) * 0.05
    Y = 7 - (frame - 302) * 0.05
    U = -2
    V = -2
    Quiver(X,Y,U,V)

# 2回転，回転行列Rotate2(時計回り)
  elif 343<= frame <=525:
    X = 5
    Y = 5
    U = -2
    V = -2
    vr = np.array([U, V])
    sind = np.sin(np.deg2rad(frame-344) * 4)
    cosd = np.cos(np.deg2rad(frame-344) * 4)
    Rotate2 = np.array([[cosd, sind],
                       [-sind, cosd]])
    vr2 = Rotate2 @ vr
    U = vr2[0]
    V = vr2[1]
    Q.set_UVC(U,V)
    Quiver(X,Y,U,V)

  elif 525< frame <=566:
    X = 5 - (frame - 525) * 0.05
    Y = 5 - (frame - 525) * 0.05
    U = -2
    V = -2
    Quiver(X,Y,U,V)

# (2, 2)を中心に180度回転＝点対称移動，回転行列Rotate3(反時計回り)
  elif 566< frame <=627:
    X = 3
    Y = 3
    U = -2
    V = -2
    vrxy = np.array([X, Y]) - np.array([2, 2])
    vruv = np.array([U, V])
    sind = np.sin(np.deg2rad(frame-566) * 3)
    cosd = np.cos(np.deg2rad(frame-566) * 3)
    Rotate3 = np.array([[cosd, -sind],
                        [sind, cosd]])
    # 原点を中心に回転させてから元に戻す
    vrxy2 = Rotate3 @ vrxy + np.array([2, 2])
    X = vrxy2[0]
    Y = vrxy2[1]
    vruv2 = Rotate3 @ vruv
    U = vruv2[0]
    V = vruv2[1]
    Q.set_UVC(U,V)
    Quiver(X,Y,U,V)

# (X, Y) = (7, 1)まで平行移動
  elif 627< frame <=648:
    X, Y = 1, 1
    U, V = 2, 2
    X = X + (frame - 627) * 0.3
    Y = Y
    Quiver(X,Y,U,V)

# (6, 4)を中心に180度回転，回転行列Rotate4(反時計回り)
  elif 648< frame <=709:
    X = 7
    Y = 1
    U = 2
    V = 2
    vrxy = np.array([X, Y]) - np.array([6, 4])
    vruv = np.array([U, V])
    sind = np.sin(np.deg2rad(frame-648) * 3)
    cosd = np.cos(np.deg2rad(frame-648) * 3)
    Rotate4 = np.array([[cosd, -sind],
                        [sind, cosd]])
    # 原点を中心に回転させてから元に戻す
    vrxy2 = Rotate4 @ vrxy + np.array([6, 4])
    X = vrxy2[0]
    Y = vrxy2[1]
    vruv2 = Rotate4 @ vruv
    U = vruv2[0]
    V = vruv2[1]
    Q.set_UVC(U,V)
    Quiver(X,Y,U,V)

# 直線y＝xに対しての線対称移動
  elif 709< frame <=740:
    # y＝xのグラフの描画
    x = np.arange(0, 10, 0.1)
    y = x
    ax.plot(x, y)
    ax.quiver(4.5, 5.5, 1, -1.25, color="black", width=0.003, headwidth=7,
              angles='xy', scale_units='xy', scale=1)
    ax.quiver(5, 7, -2, -2, color='cyan', width=0.1,  headwidth=1, headlength=0.5,
                  angles='xy', scale_units='xy', scale=1)
    if 719< frame <=740:
      X = 5
      Y = 7
      U = -2
      V = -2
      vmxy = np.array([X, Y])
      # 線対称移動用Mirror行列
      Mirror = np.array([[0, 1],
                         [1, 0]])
      # 線対称移動
      vmxy2 = Mirror @ vmxy
      X = vmxy2[0]
      Y = vmxy2[1]
      Quiver(X,Y,U,V)


# アニメーション作成
ani = animation.FuncAnimation(fig, update, fargs=(Q,X,Y), interval=50, frames=740, repeat=False)

# Display方法その1
HTML(ani.to_html5_video())

# Display方法その2(こちらにする場合は，上のHTML()は # にする)
#html = ani.to_jshtml()
#HTML(html)
```

```python

import matplotlib.patches as pat
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
from IPython.display import HTML
import math

# グラフ領域作成
fig, ax = plt.subplots(figsize=(8, 8))

# 数値設定
root3 = math.sqrt(3)

# 初期設定
def initialize():
  ax.set_xlim(0, 10)
  ax.set_ylim(0, 10)
  ax.set_xticks(np.arange(0, 10, 1))
  ax.set_yticks(np.arange(0, 10, 1))
  ax.grid()

initialize()

# 腕1ベクトル作成関数
def Quiver_A1(x,y,u,v):
  P = ax.quiver(x, y, u, v, color='green', width=0.01,  headwidth=0, headlength=0,
                headaxislength=0, angles='xy', scale_units='xy', scale=1)
  return P

# 腕2ベクトル作成関数
def Quiver_A2(X,Y,U,V):
  Q = ax.quiver(X, Y, U, V, color='cyan', width=0.01, headwidth=8, headlength=2,
                headaxislength=6, angles='xy', scale_units='xy', scale=1)
  return Q

# 回転行列(反時計回り)
def Rotate_Inv(t, z, w):
  sind = np.sin(np.deg2rad(t - z) * w)
  cosd = np.cos(np.deg2rad(t - z) * w)
  RI = np.array([[cosd, -sind],
                 [sind, cosd]])
  return RI

# 回転行列(時計回り，4倍速)
def Rotate(t, z, w):
  sind = np.sin(np.deg2rad(t - z) * w)
  cosd = np.cos(np.deg2rad(t - z) * w)
  R = np.array([[cosd, sind],
                [-sind, cosd]])
  return R

# 縮小行列(腕2だけ適用)
def Short(t, z):
  Short = np.array([[1-0.035*(t-z), 0],
                    [0, 1-0.035*(t-z)]])
  return Short


# 拡大行列
def Long(t, z):
  Long = np.array([[1+0.05*(t-z), 0],
                   [0, 1+0.05*(t-z)]])
  return Long


# 初期ベクトル設定
# 腕1(Arm1)ベクトル
x = 1
y = 4
u = 3
v = 0
P = Quiver_A1(x,y,u,v)

# 腕2(Arm2)ベクトル
X = 4
Y = 4
U = 2.5
V = 0
Q = Quiver_A2(X,Y,U,V)

plt.show()

# 腕1,2の動き設定
def Arms(t):
  if 0<= t <= 30:
    z = 0
    w = 1
    x,y,u,v,X,Y,U,V = 1,4,3,0,4,4,2.5,0

    vect = np.array([u, v])
    RI = Rotate_Inv(t, z, w)
    vect = RI @ vect
    u = vect[0]
    v = vect[1]
    P.set_UVC(u,v)
    Quiver_A1(x,y,u,v)

    S = np.array([X, Y]) - np.array([1, 4])
    S = RI @ S + np.array([1, 4])
    X = S[0]
    Y = S[1]
    Vect = np.array([U, V])
    Vect = RI @ Vect
    U = Vect[0]
    V = Vect[1]
    Q.set_UVC(U,V)
    Quiver_A2(X,Y,U,V)

  elif 30<= t <=60:
    z = 30
    w = 1
    x,y,u,v,X,Y,U,V = 1,4,1.5*root3,1.5,1+1.5*root3,5.5,1.25*root3,1.25

    vect = np.array([u, v])
    R = Rotate(t, z, w)
    vect = R @ vect
    u = vect[0]
    v = vect[1]
    P.set_UVC(u,v)
    Quiver_A1(x,y,u,v)

    S = np.array([X, Y]) - np.array([1, 4])
    S = R @ S + np.array([1, 4])
    X = S[0]
    Y = S[1]
    Vect = np.array([U, V])
    Vect = R @ Vect
    U = Vect[0]
    V = Vect[1]
    Q.set_UVC(U,V)
    Quiver_A2(X,Y,U,V)

  elif 60<= t <=75:
    z = 60
    x,y,u,v,X,Y,U,V = 1,4,3,0,4,4,2.5,0

    Quiver_A1(x,y,u,v)

    Vect = np.array([U, V])
    L = Long(t, z)
    Vect = L @ Vect
    U = Vect[0]
    V = Vect[1]
    Q.set_UVC(U,V)
    Quiver_A2(X,Y,U,V)

  elif 75<= t <=90:
    z = 75
    x,y,u,v,X,Y,U,V = 1,4,3,0,4,4,4.75,0

    Quiver_A1(x,y,u,v)

    Vect = np.array([U, V])
    S = Short(t, z)
    Vect = S @ Vect
    U = Vect[0]
    V = Vect[1]
    Q.set_UVC(U,V)
    Quiver_A2(X,Y,U,V)

  elif 90<= t <=180:
    z = 90
    x,y,u,v,X,Y,U,V = 1,4,3,0,4,4,2.5,0

    Quiver_A1(x,y,u,v)

    Vect = np.array([U, V])
    w = 4
    R = Rotate(t, z, w)
    Vect = R @ Vect
    U = Vect[0]
    V = Vect[1]
    Q.set_UVC(U,V)
    Quiver_A2(X,Y,U,V)

  elif 180<= t <= 211:
    z = 180
    w = 1
    x,y,u,v,X,Y,U,V = 1,4,3,0,4,4,2.5,0

    Quiver_A1(x,y,u,v)

    RI = Rotate_Inv(t, z, w)
    Vect = np.array([U, V])
    Vect = RI @ Vect
    U = Vect[0]
    V = Vect[1]
    Q.set_UVC(U,V)
    Quiver_A2(X,Y,U,V)
  # 腕1，2が同時に30度反時計回り
  elif 211<= t <= 241:
    z = 211
    w = 1
    x,y,u,v,X,Y,U,V = 1,4,3,0,4,4,1.25*root3,1.25

    vect = np.array([u, v])
    RI = Rotate_Inv(t, z, w)
    vect = RI @ vect
    u = vect[0]
    v = vect[1]
    P.set_UVC(u,v)
    Quiver_A1(x,y,u,v)

    S = np.array([X, Y]) - np.array([1, 4])
    S = RI @ S + np.array([1, 4])
    X = S[0]
    Y = S[1]
    Vect = np.array([U, V])
    Vect = RI @ Vect
    U = Vect[0]
    V = Vect[1]
    Q.set_UVC(U,V)
    Quiver_A2(X,Y,U,V)
  # 腕2が伸びながら330度反時計回り(10倍速)
  elif 241<= t <= 275:
    z = 241
    w = 10
    x,y,u,v,X,Y,U,V = 1,4,1.5*root3,1.5,1+1.5*root3,5.5,1.25,1.25*root3

    Quiver_A1(x,y,u,v)

    # 伸びながら330度反時計回り回転
    RI = Rotate_Inv(t, z, w)
    L = Long(t, z)
    Vect = np.array([U, V])
    Vect = RI @ L @ Vect
    U = Vect[0]
    V = Vect[1]
    Q.set_UVC(U,V)
    Quiver_A2(X,Y,U,V)

# up-date関数
def update(frame):
  ax.cla()
  initialize()

  t = frame

  Arms(t)

ani = animation.FuncAnimation(fig, update, interval=200, frames=275, repeat=False)
HTML(ani.to_html5_video())

```

ここで，上の行列のかけ算を使います

(1)の式を，1$\times$X + 1$\times$Y = 11  と書き直し

(2)の式はそのまま，2$\times$X + 4$\times$Y = 38

\

この2つの式は，行列$\begin{pmatrix} 1 & 1\\ 2 & 4  \end{pmatrix}$を

行列$\begin{pmatrix} X \\ Y  \end{pmatrix}$にかけた式と同じになっています

\

行列のかけ算の式にしてみると

$\begin{pmatrix} 1 & 1\\ 2 & 4  \end{pmatrix}$$\begin{pmatrix} X \\ Y  \end{pmatrix}$ = $\begin{pmatrix} 8 \\ 6  \end{pmatrix}$  ・・・・・(3)

になります

\

この行列のかけ算を操作して

$\begin{pmatrix} X \\ Y  \end{pmatrix}$ = $\begin{pmatrix} 3 \\ 8  \end{pmatrix}$

となれば，つるかめ算が解けたことになります

\
操作の方法の前に，下の行列のかけ算を計算してみます

$\begin{pmatrix} 1 & 0\\ 0 & 1  \end{pmatrix}$$\begin{pmatrix} X \\ Y  \end{pmatrix}$

「行$\times$列」で計算して

1$\times$X + 0$\times$Y = X

0$\times$X + 1$\times$Y = Y

です

\

すなわち

$\begin{pmatrix} 1 & 0\\ 0 & 1  \end{pmatrix}$$\begin{pmatrix} X \\ Y  \end{pmatrix}$ = $\begin{pmatrix} X \\ Y  \end{pmatrix}$

です

\

つまり，(3)の式の$\begin{pmatrix} 1 & 1\\ 2 & 4  \end{pmatrix}$が，$\begin{pmatrix} 1 & 0\\ 0 & 1  \end{pmatrix}$になるような操作をすれば

よいことが分かります

\

(3)の式の両辺に，**左から**行列Gをかけて

G$\times$$\begin{pmatrix} 1 & 1\\ 2 & 4  \end{pmatrix}$$\begin{pmatrix} X \\ Y  \end{pmatrix}$ = G$\times$$\begin{pmatrix} 11 \\ 38  \end{pmatrix}$

その結果が，

$\begin{pmatrix} 1 & 0\\ 0 & 1  \end{pmatrix}$$\begin{pmatrix} X \\ Y  \end{pmatrix}$ = G$\times$$\begin{pmatrix} 11 \\ 38  \end{pmatrix}$ ・・・・・(4)

となるためには

G$\times$$\begin{pmatrix} 1 & 1\\ 2 & 4  \end{pmatrix}$ = $\begin{pmatrix} 1 & 0\\ 0 & 1  \end{pmatrix}$

となる行列Gが分かればよいことになります

\

この行列Gは行列$\begin{pmatrix} 1 & 1\\ 2 & 4  \end{pmatrix}$の**逆行列**といいますが，

逆行列を求める公式があります

\


行列A＝$\begin{pmatrix} a & b\\ c & d  \end{pmatrix}$の逆行列は$\dfrac{1}{a \times d - b \times c}\times\begin{pmatrix} d & -b\\ -c & a  \end{pmatrix}$

\

かなり複雑なので，覚える必要はありません

\

行列$\begin{pmatrix} 1 & 1\\ 2 & 4  \end{pmatrix}$をこの公式にあてはめてみます

\

$\dfrac{1}{1 \times 4 - 1 \times 2}\times\begin{pmatrix} 4 & -1\\ -2 & 1  \end{pmatrix}$ = $\dfrac{1}{2}\times\begin{pmatrix} 4 & -1\\ -2 & 1  \end{pmatrix} 
= \begin{pmatrix} 2 & -0.5\\ -1 & 0.5  \end{pmatrix}$

\

ちなみに，(数)$\times$行列 の計算は，(数)を行列の全部の要素にかけます

\

(4)の式でGに$\begin{pmatrix} 2 & -0.5\\ -1 & 0.5  \end{pmatrix}$を代入すると

$\begin{pmatrix} 2 & -0.5\\ -1 & 0.5  \end{pmatrix}$$\begin{pmatrix} 1 & 1\\ 2 & 4  \end{pmatrix}$$\begin{pmatrix} X \\ Y  \end{pmatrix}$ = $\begin{pmatrix} 2 & -0.5\\ -1 & 0.5  \end{pmatrix}$$\begin{pmatrix} 11 \\ 38  \end{pmatrix}$

\

2行2列行列のかけ算は，「行$\times$列」の計算が4回です

また負の数のかけ算の出てきますので，ながめるだけにして下さい

\
左辺は

2$\times$1 + (-0.5)$\times$2 = 2 - 1 = 1

2$\times$1 + (-0.5)$\times$4 = 2 - 2 = 0

(-1)$\times$1 + 0.5$\times$2 = -1 + 1 = 0

(-1)$\times$1 + 0.5$\times$4 = -1 + 2 = 1

\

右辺は

2$\times$11 + (-0.5)$\times$38 = 22 - 19 = 3

(-1)$\times$11 + 0.5$\times$38 = -11 + 19 = 8

\

結局，左辺は

$\begin{pmatrix} 1 & 0\\ 0 & 1  \end{pmatrix}$$\begin{pmatrix} X \\ Y  \end{pmatrix}$ = $\begin{pmatrix} X \\ Y  \end{pmatrix}$ 

右辺は

$\begin{pmatrix} 3 \\ 8  \end{pmatrix}$

となることで答えが得られます
